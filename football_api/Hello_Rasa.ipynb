```python
import asyncio
from rasa.core.agent import Agent # Use to load your model
import nest_asyncio
```

```python
nest_asyncio.apply()
```

```python
# model_id = 'models/nlu-20240527-124040-jolly-pulse.tar.gz'
model_id = 'models/nlu-20240528-133045-gentle-graph.tar.gz'
model = Agent.load(model_path = "/content/hello_rasa/" + model_id)
```

```python
message = "Reserve me a ticket to Cairo"
pred = asyncio.run( model.parse_message(message_data = message) )
```
